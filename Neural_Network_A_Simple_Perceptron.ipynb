1.What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.
- Deep Learning is a specialized subset of machine learning that uses neural network architectures with many layers to automatically learn complex representations from large volumes of data, especially unstructured data like images, audio, or text.

- Evolution of Deep Learning
    - Deep learning evolved from traditional artificial intelligence (AI) and machine learning research. Early machine learning primarily relied on simpler algorithms capable of learning from structured, often small datasets. As computing power increased and data became more abundant, researchers began designing larger and deeper neural networks—giving rise to deep learning—which proved more effective at extracting features and learning hierarchical representations from raw, unstructured data without explicit manual feature engineering.

- Key Differences from Traditional Machine Learning
    - Data Requirements: Deep learning models require massive datasets and high computational power, while traditional machine learning works well with smaller datasets on standard hardware.

    - Feature Engineering: Traditional machine learning depends on manual feature engineering by domain experts; deep learning networks automatically extract and learn useful features directly from raw data.

    - Complexity & Performance: Deep learning excels in complex, unstructured tasks like image and speech recognition, surpassing traditional models in these areas. Traditional machine learning often achieves good results in structured, tabular tasks due to easier interpretability and faster training.

    - Architecture: Traditional models are usually shallower (few layers), while deep learning models utilize deep architectures with dozens to thousands of layers to capture intricate data patterns.

    - Automation: Deep learning models offer end-to-end learning pipelines, while traditional models require significant preprocessing and domain knowledge.

- deep learning's main advantage is its ability to automatically discover powerful representations from raw data, leading to breakthroughs in fields such as computer vision and natural language processing, while traditional machine learning remains optimal for less complex use-cases with limited data and a need for model interpretability.


2.Explain the basic architecture and functioning of a Perceptron. What are its
limitations?
- A Perceptron is the simplest form of a neural network, designed to help computers make decisions for problems where only two outcomes are possible, like yes/no or true/false.

- How a Perceptron Works
  - The perceptron receives several input values (which can be numbers representing features).

  - Each input is multiplied by a weight, showing how important that input is.

  - These multiplied inputs are all added together along with a bias (a number that helps adjust the output for better accuracy).

  - The result goes through an activation function, usually a step function, which turns the result into either a 1 (yes/true) or 0 (no/false).

  - If the answer is above a set threshold, the perceptron outputs 1; otherwise, it outputs 0.

- Example (in simple steps)
  - Take each input and multiply by its weight.

  - Add all these results plus the bias.

  - Check if the total is above the threshold. If yes, output 1. If not, output 0.

- Limitations of a Perceptron
  - Only for simple problems: It can only solve tasks where the answers can be separated by a straight line (linearly separable problems).

  - Binary decision: It only gives answers as 0 or 1, so it can't handle problems that need more than two possible outcomes or more complex relationships.

  - Not for complex tasks: It cannot learn non-linear patterns or solve challenging problems like recognizing handwritten digits on its own.

  - Single-layer only: One perceptron alone can't extract deeper, complicated patterns from input data; multiple layers (multi-layer perceptrons) are needed for those tasks

3.Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions
- In neural networks, the activation function plays a crucial role. Its main purpose is to decide whether a neuron should be activated or not based on its input. This means transforming the sum of the weighted inputs a neuron receives into an output that can either be passed on to the next layer or used as the final prediction result. Activation functions are important because they introduce non-linearity to the network, enabling it to learn and model complex patterns in data. Without activation functions, regardless of how many layers a neural network has, it would behave like a simple linear model, greatly limiting its ability to solve real-world problems that often involve non-linear relationships.

- Why Activation Functions are Needed
  - Think of a neuron like a gatekeeper that receives incoming signals. The activation function acts like the decision maker, determining if those signals are strong or important enough to pass on. This decision-making mimics the behavior of neurons in the human brain, which either fire or stay inactive depending on the intensity of the incoming signals. Moreover, activation functions allow the neural network to stack multiple layers to process data hierarchically and extract meaningful features at each stage. Without them, the combined effect of multiple layers would still be just a simple linear transformation, making deep networks ineffective.

- Comparing Sigmoid, ReLU, and Tanh Activation Functions
  - Sigmoid Function: The sigmoid activation squashes any input value into a range between 0 and 1. This makes it a good choice for binary classification problems because its output can be interpreted as a probability. However, it has a downside — when inputs are very high or very low, the slope flattens, causing the gradient to vanish and slowing down training. It is mostly used in the output layer of some neural networks but is less common in hidden layers due to these issues.

  - ReLU (Rectified Linear Unit): ReLU activates the neuron only if the input is positive, outputting zero for any negative input. This simple behavior makes it computationally efficient and helps networks learn much faster. ReLU also reduces the vanishing gradient problem that plagues sigmoid functions. However, ReLU can sometimes lead to neurons "dying," meaning some neurons might output zero for all inputs and stop learning altogether.

  - Tanh (Hyperbolic Tangent): The tanh function scales input values to lie between -1 and 1, offering a zero-centered output (unlike sigmoid, which is between 0 and 1). This zero-centered nature often makes tanh more effective than sigmoid in hidden layers. However, it can still suffer from the vanishing gradient problem when input values become large or small.
- activation functions help neural networks learn complex data relationships by introducing essential non-linearity. Sigmoid and tanh are smooth and suitable for some tasks but can suffer from vanishing gradients, while ReLU is faster and often preferred in deeper networks despite the risk of dead neurons. Choosing the right activation function depends on the problem at hand and the network architecture

4.What is the difference between Loss function and Cost function in neural networks? Provide examples.
- In neural networks and machine learning, the terms "loss function" and "cost function" are closely related but have distinct meanings that are important to understand.

- Loss Function
  - The loss function measures the error for a single training example. It calculates how far off the model's prediction is from the true value or label for that one data point. Essentially, it gives a numeric value representing the penalty or "loss" for the prediction being inaccurate. For every example in the training set, there is a corresponding loss value.

  - For example, if predicting house prices, the loss function for one house could be the absolute difference between the actual price and the predicted price. If the actual price is $300,000 and the model predicts $320,000, the loss from this example would be 20,000. This loss quantifies how wrong the model was for that specific point.

- Cost Function
  - While the loss function deals with individual examples, the cost function looks at the overall error across the entire dataset or a batch of samples. It is the average (or sometimes sum) of all the individual loss values computed from every training example. The cost function provides a single value representing how well the model is performing on the whole dataset.

  - For instance, in the house price prediction example, the cost function could be the mean absolute error across all houses in the dataset. It averages the losses from all predictions to get a general measure of how accurate or inaccurate the model is overall. The goal during training is to minimize this cost function.


5.What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.
- Optimizers in neural networks play a fundamental role in training models by systematically adjusting the network's parameters, such as weights and biases, to minimize errors in prediction. The process of training a neural network involves repeatedly making predictions, comparing them to true values using a loss function, and then using an optimizer to update the parameters so that the prediction errors become as small as possible in subsequent iterations. Without optimizers, neural networks would lack a methodical way to learn from data and improve their accuracy over time.

- Imagine trying to find the lowest point in a foggy mountain landscape — you must take cautious steps downhill, adjusting your direction based on the slope you feel beneath your feet. Similarly, optimizers help the neural network "descend" the landscape of the loss function by tweaking parameters toward the minimum error point. The landscape, or loss function, measures how far off the model's predictions are, and the optimizer guides the parameter updates that reduce this error.

- Gradient Descent
  - Gradient Descent is the most foundational optimization algorithm. It works by calculating the gradient (or slope) of the loss function with respect to each parameter and then taking steps in the opposite direction of the gradient to reduce error. This method uses the entire training dataset to compute the gradient at each step, which can be computationally expensive for large datasets. Gradient Descent has a fixed learning rate that determines how big a step is taken each time it updates the parameters. While it's simple and easy to implement, it can converge slowly and may get stuck in flat regions or local minima.

- Adam (Adaptive Moment Estimation)
  - Adam is a sophisticated and widely used optimizer that builds upon Gradient Descent by incorporating adaptive learning rates and momentum. It maintains moving averages of both the gradients and their squared values, which helps the algorithm adaptively adjust the learning rate for each parameter individually. This allows Adam to handle noisy or sparse gradients well and speeds up convergence. Because of its robustness and efficiency, Adam is often the default choice for many deep learning models.

- RMSprop (Root Mean Square Propagation)
  - RMSprop is another adaptive learning rate optimizer that adjusts the learning rate for each parameter based on a moving average of recent squared gradients. This mechanism helps stabilize the training by preventing large oscillations, especially in scenarios with noisy or non-stationary data like recurrent neural networks. RMSprop ensures that parameters with large gradients get smaller updates, and vice versa, which improves performance and convergence speed over basic Gradient Descent.

```python
#6. Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, n_iters=10):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        # Initialize weights and bias to zeros
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Training
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation(linear_output)

                # Perceptron update rule
                update = self.lr * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def activation(self, x):
        # Step activation function
        return 1 if x >= 0 else 0

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = np.array([self.activation(i) for i in linear_output])
        return y_predicted

# Logical AND gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # AND output

# Create perceptron instance
p = Perceptron(learning_rate=0.1, n_iters=10)

# Train perceptron
p.fit(X, y)

# Test trained perceptron
predictions = p.predict(X)
print("Input data:\n", X)
print("Predictions:", predictions)
print("Expected:   ", y)
```

```python
#7.Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Tanh activation function
def tanh(x):
    return np.tanh(x)

# Input range
x = np.linspace(-10, 10, 400)

# Compute activation outputs
sigmoid_output = sigmoid(x)
relu_output = relu(x)
tanh_output = tanh(x)

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid_output, label="Sigmoid", color='blue')
plt.plot(x, relu_output, label="ReLU", color='green')
plt.plot(x, tanh_output, label="Tanh", color='red')

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.show()
```

```python
#8. Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.datasets import mnist

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to the range [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Build the neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),        # Flatten 28x28 images into 784-dimensional vector
    Dense(128, activation='relu'),        # First hidden layer with 128 neurons and ReLU activation
    Dense(64, activation='relu'),         # Second hidden layer with 64 neurons and ReLU activation
    Dense(10, activation='softmax')       # Output layer with 10 neurons (one per digit) and softmax activation
])

# Compile the model with optimizer, loss function, and metrics
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, verbose=1)

# Print training accuracy after the final epoch
train_accuracy = history.history['accuracy'][-1]
print(f"Training accuracy after last epoch: {train_accuracy*100:.2f}%")
```

```python
#9.Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalize pixel values to [0,1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Build a simple neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),        # Flatten 28x28 images
    Dense(128, activation='relu'),        # Hidden layer with ReLU activation
    Dense(64, activation='relu'),         # Another hidden layer
    Dense(10, activation='softmax')       # Output layer for 10 classes
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model and save training history
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Plot loss curves
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot accuracy curves
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='orange')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.show()
```

10.You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:
● How would you design a deep learning model (perceptron or multilayer NN)?
● Which activation function and loss function would you use, and why?
● How would you train and evaluate the model, considering class imbalance?
● Which optimizer would be suitable, and how would you prevent overfitting?
- For a bank's project to automatically detect fraudulent transactions using a large, imbalanced dataset of structured features, the real-time data science workflow would encompass the following elements:

- Designing the Deep Learning Model
  - Model Type: Use a multilayer neural network (deep neural network, DNN) rather than a simple perceptron because fraud detection involves complex, non-linear interactions between features like transaction amount, merchant IDs, and customer locations.

  - Architecture: Design with multiple dense (fully connected) layers, e.g., three hidden layers with decreasing units (e.g., 128 → 64 → 32 nodes) for hierarchical feature learning. Include dropout layers after each dense layer to reduce overfitting.

  - Output Layer: Use a single neuron with a sigmoid activation function to output the probability a transaction is fraudulent (values between 0 and 1).

- Activation and Loss Functions
  - Activation Functions: Use ReLU (Rectified Linear Unit) for hidden layers because it is computationally efficient, helps in mitigating vanishing gradient issues, and accelerates convergence. Use sigmoid for the output layer since this is a binary classification problem.

  - Loss Function: Use binary cross-entropy loss because the target is binary (fraud or legitimate), and this loss function is suitable to measure the distance between predicted probabilities and true labels in such cases.

- Training and Evaluation with Class Imbalance
  - Handling Imbalance: Because fraudulent transactions are rare, use techniques like:

    - Class weighting: Modify the loss function to penalize mistakes on the minority class (fraud) more heavily.

    - Resampling methods: Oversample the minority class (e.g., SMOTE) or undersample the majority class to balance the dataset.

  - Evaluation Metrics: Use precision, recall, F1-score, and area under the ROC curve (AUC) rather than simple accuracy, since accuracy can be misleading on imbalanced datasets. Focus especially on recall (sensitivity) for fraud detection to minimize false negatives.

  - Validation: Use stratified k-fold cross-validation or hold out a validation set representative of the imbalance.

- Optimizer and Overfitting Prevention
  - Optimizer: Adam optimizer is a good choice because it adapts learning rates individually for each parameter and converges quickly in many practical settings.

  - Regularization: Use dropout layers and possibly L2 regularization to prevent overfitting.

  - Early Stopping: Monitor validation loss and stop training if it stops improving to avoid overfitting.

  - Batch Normalization: Optionally incorporate batch normalization layers to stabilize and speed up training.
